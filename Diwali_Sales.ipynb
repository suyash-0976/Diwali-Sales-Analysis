***Data Analysis***

```python
df.columns
```

```python
ax = sns.countplot(x= 'Gender', data=df)
# taki wo bar chart mai dikh paye
for bars in ax.containers:
  ax.bar_label(bars)
# taki values bhi aye paye
```

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
# taki group by ho paye amount ki baisi pe aur gender ke upar aur wo sort ho
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x = 'Gender', y='Amount', data = sales_gen)
```

as shown above the most of the buyesrs are females and even the purchasing power of them are greater as compared to women

***Age***

```python
sns.countplot(data = df, x = 'Age Group', hue= 'Gender')
# age pata chalne ke liye aur hue taki gender
```

```python
ax = sns.countplot(data = df, x= 'Age Group', hue = 'Gender')

for bars in ax.containers:
  ax.bar_label(bars)
```

```python
# total amount vs age group
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Age Group', y = 'Amount',data= sales_age)
```

 as shown in above grpah we can see that most of the buyers are between 25-35 that too wome

***State***

```python
# total no. of orders kitne hai top 12 states mai se
sales_state = df.groupby('State', as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(12)

sns.set(rc={'figure.figsize':(25,10)}) # size

sns.barplot(data = sales_state, x = 'State', y = 'Orders')
```

```python
# taki pata chale total amount/sales 12 states mai se
sales_state = df.groupby('State', as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(12)

sns.set(rc={'figure.figsize':(20,10)}) # size
sns.barplot(data = sales_state, x = 'State', y = 'Amount')
```

ye pata chala ki top byuyres up ,mh,karnataka but kerla ki buying power kam hai haryana bihar aur gujrat se although usse zyda oredrs hai wo 3 states se

***Martial Status***

```python
ax = sns.countplot(data = df, x ='Marital_Status')

sns.set(rc={'figure.figsize':(5,5)})
for bars in ax.containers:
  ax.bar_label(bars)
# kons zyada mareid ya unmaried
```

```python
# kon zyda spend kar raha
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(5,5)}) # size
sns.barplot(data = sales_state, x = 'Marital_Status', y = 'Amount', hue='Gender')
```

Most shopping is done by married women  

***Occupation***

```python
sns.set(rc={'figure.figsize' :(25,5)})
ax = sns.countplot(data = df, x ='Occupation')

for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_state= df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(22,5)}) # size
sns.barplot(data = sales_state, x = 'Occupation', y = 'Amount')
```

as we can see most purchasing is done by IT Sector than HealthCare Aviatin

***Product Category***

```python
sns.set(rc={'figure.figsize' :(25,5)})
ax = sns.countplot(data = df, x ='Product_Category')

for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_state= df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(9)

sns.set(rc={'figure.figsize':(22,5)}) # size
sns.barplot(data = sales_state, x = 'Product_Category', y = 'Amount')
```

as we can see Food is most purhcase than clothing and all

```python
# most soled prodcucts
sales_state= df.groupby(['Product_ID'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(9)

sns.set(rc={'figure.figsize':(22,5)}) # size
sns.barplot(data = sales_state, x = 'Product_ID', y = 'Amount')
```
